```python
# Import required modules
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")

# Import DiagnosisForth directly to avoid import issues
import sys
sys.path.insert(0, str(Path.cwd().parent))

# Now import DiagnosisForth modules
from DiagnosisForth.readers.pdb_reader import read_multiframe_pdb

# Read multi-frame PDB file using new function name
frames = read_multiframe_pdb('./data/complex_pose_0.pdb')
```

```python
# Import analyzer functions
from DiagnosisForth.interactions.analyzer import analyze_protein_ligand_interactions_from_mol

# Analyze interactions for a specific frame using new function name
idx = 98
interactions_by_site = analyze_protein_ligand_interactions_from_mol(frames[idx])
index_of_selected_site = 0
selected_site = list(interactions_by_site.keys())[index_of_selected_site]
```

```python
# Convert hydrogen bond interactions to DataFrame
from DiagnosisForth.interactions.analyzer import interactions_to_dataframe

hbond_df = interactions_to_dataframe(
    binding_site=interactions_by_site[selected_site], 
    interaction_type="hbond"
)
hbond_df
```

```python
# Visualize pharmacophore features
from DiagnosisForth.viewers.utils import visualize_pharmacophore_from_mol

# Visualize pharmacophore features using new function name
view, tables, pharmacophore_df = visualize_pharmacophore_from_mol(
    pdb_mol=frames[idx],
    smiles='O=C(O)CCCN1CC(Oc2c1cccc2NC(=O)c1ccc(cc1)OCCCCCc1ccccc1)C(=O)O',
    ligand_name='UNK'
)
view.show()
```
